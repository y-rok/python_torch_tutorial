```python
# https://wikidocs.net/92071

import matplotlib.pyplot as plt

plt.plot([1,2,3,4]) # y의 값들 / x의 값들 입력 안하면 자동으로 0,1,2,3으로 설정됨
plt.ylabel('y-label')
plt.show()

plt.plot([1,2,3,4],[1,4,9,16],'ro') # x, y 값들 / ro - 빨간색 원형 마커
plt.show()
```

```python
# https://wikidocs.net/4761
# figure와 sub plot을 활용하여 한 화면에 여러개의 그림 표현 가능

fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,1,1) # 2행 1열 중 1번째 그림 
ax2 = fig.add_subplot(2,1,2) # 2행 1열 중 2번째 그림
# 위의 코드는 아래의 코드로도 대체 가능 
# ax1 = fig.add_subplot(211) # 2행 1열 중 1번째 그림
# ax2 = fig.add_subplot(212) # 2행 1열 중 2번째 그림

x = range(0,10)
y = [v*v for v in x]
ax1.plot(x,y,"r--")
ax2.bar(x,y) # 막대 그래프 형태로 그림

plt.show()


```
